```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("bike_sales_india.csv")
```

```python
print(df.shape)
df.head()
```

```python
df.columns
```

```python
df.info()
print(f"Total null values present in the dataset are : {df.isnull().sum().sum()}")
```

```python
df["Brand"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90 )
plt.title("Kawasaki has the highest brand market share")
plt.show()
```

```python

plt.hist(df["Price (INR)"], bins=20, edgecolor="black")
plt.title("Uniform Distribution of Bike Prices")
plt.xlabel("Price (INR)")
plt.ylabel("Number of Bikes")
plt.show()
```

```python
data=df.groupby("Brand")["Price (INR)"].sum()
top_brand = data.idxmax()
plt.bar(data.index, data.values)
plt.title(f"{top_brand} generated the highest total sales value")
plt.xlabel("Brand")
plt.ylabel("Total Sales Value (INR)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
data = df.groupby("Model")["Price (INR)"].sum().sort_values(ascending=False).head(10)
plt.bar(data.index,data.values)
plt.xticks(rotation=45)
plt.title("Top 10 Models by Sales Value")
plt.show()
```

```python
data=df.groupby("Brand")["Price (INR)"].mean().sort_values(ascending=False)
top_brand = data.idxmax()
plt.bar(data.index, data.values)
plt.title(f"{top_brand}  generated the highest average revenue")
plt.xlabel("Brand")
plt.ylabel("Total Sales Value (INR)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
data=df.groupby("State")["Price (INR)"].sum().sort_values(ascending=False)
top_state = data.idxmax()
plt.barh(data.index, data.values)
plt.title(f"{top_state} generates highest total sales value from bikes")
plt.xlabel("State")
plt.xlabel("Total Sales Value (INR)")   
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
data=df.groupby("Year of Manufacture")["Price (INR)"].sum().sort_values(ascending=False)
top_year = data.idxmax()
plt.plot(data.index.astype(str), data.values, marker="o", color="red", linewidth=3)
plt.title(f"{top_year} recorded the highest sales value of bikes")
plt.xlabel("Year of Manfacture")
plt.ylabel("Total Sales Value (INR)")   
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
data=df["Owner Type"].value_counts()
top_owner = data.idxmax()
plt.bar(data.index, data.values)
plt.title(f"{top_owner} Owner Type Has the Highest Number of Bikes")
plt.xlabel("Owner Type")
plt.ylabel("Number of Bikes")
plt.tight_layout()
plt.show()
```

```python
df["Fuel Type"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90 )
plt.title("Hybrid and Petrol-powered models are widely used")
plt.show()
```

```python
df["Seller Type"].value_counts().plot(kind="bar")
plt.xticks(rotation=0)
plt.title("Dealer Generates the highest sales of bikes")
plt.show()
```
